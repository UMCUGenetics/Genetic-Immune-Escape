# Function

```python
def run_loh_test(d_data,q):
    d=q.groupby(["SampleId","Allele"]).agg({"Peptide":"count"}).to_dict()["Peptide"]
    for allele in ["HLA-A","HLA-B","HLA-C"]:
        for sample in d_data[allele]:
            loss,kept = d_data[allele][sample]
            loss_neo=d.get((sample,loss),0)
            kept_neo=d.get((sample,kept),0)
            if (loss_neo + kept_neo) == 0 :
                continue
            ratio_normalized = (loss_neo) / (loss_neo + kept_neo)
            if kept_neo == 0:
                kept_neo=1
            ratio = loss_neo / (kept_neo)
            if loss_neo == 0:
                loss_neo = 1
            ratio_alt = kept_neo / loss_neo
            l.append([sample,allele,loss_neo,kept_neo,ratio_normalized,1-ratio_normalized,ratio,ratio_alt,loss,kept])
    df_a = pd.DataFrame(l,columns=["sampleId","gene","loss_neo","kept_neo","norm_ratio","norm_alt_ratio","ratio_basic","ratio_alt_basic","loss_allele","kept_allele"]) 
    return df_a
def run_loh_test_cumulative(d_data,q):
    l=[]
    
    for sample in loh["sampleId"].values:
        loss_neo,kept_neo=0,0
        try:
            d=q.loc[[sample]]
        except:
            continue
        for allele in ["HLA-A","HLA-B","HLA-C"]:
            if not(sample in d_data[allele]): # allele not lost!
                continue 
            loss,kept = d_data[allele][sample]
            if not(sample in q.index):
                continue    
            loss_neo=+d[(d["Allele"]==loss)].shape[0]
            kept_neo=+d[(d["Allele"]==kept)].shape[0]
        if (loss_neo + kept_neo) == 0 :
            continue
        ratio_normalized = (loss_neo) / (loss_neo + kept_neo)

        if kept_neo == 0:
            kept_neo=1
        ratio = loss_neo / (kept_neo)
        if loss_neo == 0:
            loss_neo = 1
        ratio_alt = kept_neo / loss_neo
        l.append([sample,loss_neo,kept_neo,ratio_normalized,1-ratio_normalized,ratio,ratio_alt])
    df_a = pd.DataFrame(l,columns=["sampleId","loss_neo","kept_neo","norm_ratio","norm_alt_ratio","ratio_basic","ratio_alt_basic"]) 
    return df_a
def run_loh_test_ttype(d_data,q,ttype):
    l=[]
    d=q.groupby(["SampleId","Allele"]).agg({"Peptide":"count"}).to_dict()["Peptide"]
    for allele in ["HLA-A","HLA-B","HLA-C"]:
        for sample in d_data[allele]:
            if not(sample in samples_ttype[ttype]):
                continue
            loss,kept = d_data[allele][sample]
            loss_neo=d.get((sample,loss),0)
            kept_neo=d.get((sample,kept),0)
            if (loss_neo + kept_neo) == 0 :
                continue
            ratio_normalized = (loss_neo) / (loss_neo + kept_neo)
            if kept_neo == 0:
                kept_neo=1
            ratio = loss_neo / (kept_neo)
            if loss_neo == 0:
                loss_neo = 1
            ratio_alt = kept_neo / loss_neo
            l.append([sample,allele,loss_neo,kept_neo,ratio_normalized,1-ratio_normalized,ratio,ratio_alt])
    df_a = pd.DataFrame(l,columns=["sampleId","gene","loss_neo","kept_neo","norm_ratio","norm_alt_ratio","ratio_basic","ratio_alt_basic"]) 
    return df_a
def plot_boxplot(df_a,title,name):
    e=df_a.melt(id_vars=["gene"],value_vars=["norm_ratio","norm_alt_ratio"],var_name=["lost_allele"])
    fig,ax = plt.subplots(figsize=(5,5))
    sns.boxenplot(data=e,x="lost_allele",y="value",ax=ax,palette={"norm_ratio":"green","norm_alt_ratio":"grey"},showfliers=False)
    sns.stripplot(data=e,x="lost_allele",y="value",ax=ax,palette={"norm_ratio":"green","norm_alt_ratio":"grey"},split=True,alpha=0.1)
    add_stat_annotation(ax=ax, data=e, x="lost_allele", y="value",box_pairs=[("norm_ratio","norm_alt_ratio")],comparisons_correction=None,  test='Mann-Whitney', text_format='full',  loc='inside', verbose=0)
    #ax.axhline(y=0.5,ls="--",color="grey")
    ax.set_xticklabels(["lost allele","kept_allele"],fontsize=12)
    ax.set_xlabel("")
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_title(title,fontsize=14)
    ax.set_ylabel("number neo. allele 1 / (total predicted neo.) ",fontsize=12)
    ax.annotate(xy=(0.15,0.0),s=f"N={e[e['lost_allele']=='norm_ratio'].shape[0]} alleles ",fontsize=14)
    plt.savefig(f'plots/LOH_neo_effect_{name}.pdf', dpi=800,bbox_inches="tight")
    
def plot_boxplot_gene(df_a,title,name):
    e=df_a.melt(id_vars=["gene"],value_vars=["norm_ratio","norm_alt_ratio"],var_name=["lost_allele"])
    fig,ax = plt.subplots(figsize=(5,5))
    sns.boxenplot(data=e,hue="lost_allele",y="value",ax=ax,palette={"norm_ratio":"green","norm_alt_ratio":"grey"},showfliers=False,x="gene")
    sns.stripplot(data=e,hue="lost_allele",y="value",ax=ax,palette={"norm_ratio":"green","norm_alt_ratio":"grey"},x="gene",split=True,alpha=0.1)
    add_stat_annotation(ax=ax, data=e, hue="lost_allele", y="value",x="gene",box_pairs=[(("HLA-A","norm_ratio"),("HLA-A","norm_alt_ratio")),
                                                                                       (("HLA-B","norm_ratio"),("HLA-B","norm_alt_ratio")),
                                                                                       (("HLA-C","norm_ratio"),("HLA-C","norm_alt_ratio"))],comparisons_correction=None,  test='Mann-Whitney', text_format='simple',  loc='inside',verbose=0)
    #ax.axhline(y=0.5,ls="--",color="grey")

    ax.set_xlabel("")
    ax.spines['right'].set_visible(False)
    ax.set_ylim(0.0,1.0)
    ax.spines['top'].set_visible(False)
    ax.set_title("Likelihood < 2%, ExpLikelihood < 2%")
    ax.set_ylabel("# neo. allele 1 / (total neo. alleles 1/2) ")
    ax.legend([])
    ax.set_title(title,fontsize=14)
    ax.set_ylabel("number neo. allele 1 / (total predicted neo.) ",fontsize=12)
    ax.annotate(xy=(0.0,0.0),s=f"N={e[(e['lost_allele']=='norm_ratio')&(e['gene']=='HLA-A')].shape[0]} ",fontsize=14)
    ax.annotate(xy=(1.0,0.0),s=f"N={e[(e['lost_allele']=='norm_ratio')&(e['gene']=='HLA-B')].shape[0]} ",fontsize=14)
    ax.annotate(xy=(2.0,0.0),s=f"N={e[(e['lost_allele']=='norm_ratio')&(e['gene']=='HLA-C')].shape[0]} ",fontsize=14)
    plt.savefig(f'plots/LOH_neo_effect_gene_{name}.pdf', dpi=800,bbox_inches="tight")
```

# Analysis of ratios

```python
def run_loh_test_ratio(d_data,q):
    l=[]
    d=q.groupby(["SampleId","Allele"]).agg({"Peptide":"count"}).to_dict()["Peptide"]
    for allele in ["HLA-A","HLA-B","HLA-C"]:
        for sample in d_data[allele]:
            loss,kept = d_data[allele][sample]
            loss_neo=d.get((sample,loss),0)
            kept_neo=d.get((sample,kept),0)
            if loss_neo + kept_neo ==0:
                continue
            is_a1=np.random.randint(low=0,high=2)
            if  is_a1 == 0:
                loh=True
                ratio_normalized = (loss_neo) / (loss_neo + kept_neo)
            else:
                loh=False
                ratio_normalized = (kept_neo) / (loss_neo + kept_neo)
                
            l.append([sample,allele,loss_neo,kept_neo,ratio_normalized,1-ratio_normalized,loh,np.random.randint(low=0,high=2)==0])
    df_a = pd.DataFrame(l,columns=["sampleId","gene","loss_neo","kept_neo","norm_ratio","norm_alt_ratio","loh","loh_random"]) 
    return df_a
thresholds=list([0.35,0.40,0.45,0.5,0.55,0.6,0.65,1.0])
thresholds_cat = ["[0.0,0.35)","[0.35,0.4)","[0.4,0.45)","[0.45,0.5)","[0.5,0.55)","[0.55,0.6)","[0.6,0.65)","[0.65,1.0)"]
def get_bin(k):
    for i,v in enumerate(thresholds[0:]):
        if k < v:
            return i
    return i
def compute_prevalence_mean(k):
    return  np.sum(k)/len(k)
```

### Multiple iterations

```python
l=[]
q=df[(df["LikelihoodRank"]<0.02)&(df["ExpLikelihoodRank"]<0.02)]
for N in tqdm(range(100)):
    df_ = run_loh_test_ratio(d_data_full,q.sample(frac=0.75,random_state=N))
    df_["iteration"] = N
    l.append(df_)
    
df_a = pd.concat(l)
df_a.to_csv("data/raw_data_loh_randomizations_hmf.tsv.gz",sep="\t",index=False,compression="gzip")
```
