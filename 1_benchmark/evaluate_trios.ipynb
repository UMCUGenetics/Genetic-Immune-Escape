```python
import pandas as pd
pd.options.display.max_columns=100
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
df_data= pd.read_csv("data/lilac_platinum_yoruba_trios.csv") # will be part of supp. table 2
```

### By patient and by total number of alleles

```python
# number of alleles
total_alleles=df_data.shape[0]
correct_alleles=df_data[(df_data["is_validated"])].shape[0]
```

```python
fig,ax = plt.subplots(figsize=(3,1.5))
yticks_labels = []
yticks = []
y=0
yticks.append(y)
yticks_labels.append("Number of tested HLA-I alleles")
ax.barh(y=y,width=total_alleles,color="grey",edgecolor="black",linewidth=1.5,height=0.7)
ax.annotate(xy=(total_alleles+5,y),s=str(total_alleles),fontsize=16 )

y-=1
yticks_labels.append("Correct HLA-I typing")
yticks.append(y)
ax.barh(y=y,width=correct_alleles,color="#4daf4a",edgecolor="black",linewidth=1.5,height=0.7)
ax.annotate(xy=(correct_alleles+5,y),s=str(correct_alleles) + "  ({0:.2%})".format(correct_alleles/total_alleles),fontsize=16)

ax.set_yticks(yticks)
ax.set_yticklabels(yticks_labels)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Number of HLA-I alleles",fontsize=16)
ax.set_title("LILAC's HLA-I typing on Platinum and Yoruban trios",fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_xlim(0,60)
plt.savefig(f'../results/figures/figure1/platinum_yoruban_trios.pdf', dpi=800,bbox_inches="tight")
```
